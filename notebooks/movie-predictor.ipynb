```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
```

```python
movies_df = pd.read_csv('../dataset/final/movies_final.csv')
movies_df.head()
```

```python
movies_df = pd.read_csv('../dataset/final/movies_final_v32.csv')
movies_df.head()
```

```python
train_arr = np.array([eval(entry) for entry in movies_df['labels']])
train_arr
```

```python
model = NearestNeighbors(n_neighbors=10, metric='cosine')
model.fit(train_arr)
```

```python
import pickle
pickle.dump(model, open('cross-recommender-model-v32.pkl', 'wb'))
```
